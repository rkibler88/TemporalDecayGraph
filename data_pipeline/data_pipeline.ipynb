this here is the data pipeline for turning the data into tuples!

```python
#load packages
import os
import numpy as np
import mne
import pandas as pd
from scipy.signal import coherence
from itertools import combinations
```

```python
# loading data from:https://physionet.org/content/eegmat/1.0.0/
df = mne.io.read_raw_edf("Subject00_1.edf", preload=True, encoding='latin1')
df = df.to_data_frame()
```

```python
#the data is in 500hz so we make fs 500
fs = 500
#window size is 500 samples bc 500Hz
window_size = fs * 1
#rows // 500 = ~355 seconds (windows)
n_windows = len(df) // window_size

#get all channels as long as they are eeg channels
channels = []
for col in df.columns:
    if col.startswith('EEG'):
        channels.append(col)


# window loop to find coherence of pairs per window
output_list = []
for w in range(n_windows):
    print(f"Processing window {w} of {n_windows}")
    start = w*window_size
    end = start+window_size
    window_data=df.iloc[start:end]
    for i in range(len(channels)):
        for j in range(i+1, len(channels)):
            channel1 = channels[i]
            channel2 = channels[j]
            freqs, coh = coherence(window_data[channel1].values, window_data[channel2].values, fs=500.0)
            output = {'epoch_id': w, 'channel_i': channel1, 'channel_j': channel2, 'coherence_value': coh.mean()}
            output_list.append(output)

output_df = pd.DataFrame(output_list)
```
